#### Q2.
Sobre os dados de copa do mundo, qual a distribuição de público presente nos jogos? Isso pode ser respondido com um histograma com os dados da coluna `comparecimento`.  

Lembre-se que alguns jogos estavam com público 0 incorretamente, que tal remover essas ocorrências para não atrapalhar sua visualzação?

Você deve implementar essa visualização nas três bibliotecas que vimos:
* Matplotlib
* Seaborn
* Plotly

Garanta que o gráfico tenha pelo menos os atributos de título e rótulos de dimensão.

````python
#### Solução com matplotlib

import matplotlib.pyplot as plt

# Remover jogos com público igual a 0
wc_filtrado = wc[wc['comparecimento'] > 0]

# Criar o histograma
plt.figure(figsize=(10,6))  # Define o tamanho do gráfico
plt.hist(wc_filtrado['comparecimento'], bins=30, color='skyblue', edgecolor='black')

# Adicionar título e rótulos
plt.title('Distribuição de Público nos Jogos de Copa do Mundo', fontsize=16)
plt.xlabel('Público presente', fontsize=14)
plt.ylabel('Quantidade de jogos', fontsize=14)

# Mostrar o gráfico
plt.show()
````

````python
#### solução com seaborn
import seaborn as sns
import matplotlib.pyplot as plt

# Remover jogos com comparecimento 0
wc_filtrado = wc[wc['comparecimento'] > 0]

# Criar o histograma
plt.figure(figsize=(10,6))
sns.histplot(data=wc_filtrado, x='comparecimento', bins=30, color='purple', edgecolor='black')

# Adicionar título e rótulos
plt.title('Distribuição de Público nos Jogos de Copa do Mundo', fontsize=16)
plt.xlabel('Público presente', fontsize=14)
plt.ylabel('Quantidade de jogos', fontsize=14)

# Mostrar
plt.show()
````

````python
#### solução com plotly
import plotly.express as px

# Remover jogos com comparecimento 0
wc_filtrado = wc[wc['comparecimento'] > 0]

# Criar o histograma
fig = px.histogram(
    wc_filtrado,
    x='comparecimento',
    nbins=30,
    title='Distribuição de Público nos Jogos de Copa do Mundo',
    labels={'comparecimento': 'Público presente'},
)

# Personalizar layout
fig.update_layout(
    xaxis_title='Público presente',
    yaxis_title='Quantidade de jogos',
    title_x=0.5,
    template='plotly_dark'
)

# Mostrar o gráfico
fig.show()
````

#### Q3.

Apresente um gráfico de dispersão (scatter) dos atributos `gols_1` e `gols_2`. Isso representa a relação entre gols feitos e gols tomados por jogo. Há alguma relação interessante entre esses atributos?

Para facilitar a visualização dos dados (já que tem muitos placares repetidos), aplique uma leve distorção aos dados para que cada ponto esteja deslocado aleatoriamente de seu valor original. Código apresentado a seguir
```python
gols = wc[['gols_1', 'gols_2']] * np.random.random((len(wc),2))
```

Você deve implementar essa visualização nas três bibliotecas que vimos:
* Matplotlib
* Seaborn

Garanta que o gráfico tenha pelo menos os atributos de título e rótulos de dimensão.

````python
import numpy as np
import pandas as pd

# Lendo o wc_formatado.csv
wc = pd.read_csv('wc_formatado.csv')

# Cria o DataFrame com distorção aleatória
gols = wc[['gols_1', 'gols_2']] * np.random.random((len(wc), 2))
````

````python
#### solução com matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

# Gráfico
plt.figure(figsize=(8,6))
plt.scatter(gols['gols_1'], gols['gols_2'], alpha=0.5)
plt.title('Dispersão de Gols Feitos vs Gols Sofridos (Matplotlib)', fontsize=14)
plt.xlabel('Gols Marcados (gols_1)')
plt.ylabel('Gols Sofridos (gols_2)')
plt.grid(True)
plt.show()
````

````python
#### solução com seaborn
import seaborn as sns

sns.set_theme(style="whitegrid")
# Gráfico
plt.figure(figsize=(8,6))
sns.scatterplot(x=gols['gols_1'], y=gols['gols_2'])
plt.title('Dispersão de Gols Feitos vs Gols Sofridos (Seaborn)', fontsize=14)
plt.xlabel('Gols Marcados (gols_1)')
plt.ylabel('Gols Sofridos (gols_2)')
plt.grid(True)
plt.show()
````

````python
#### solução com plotly
import plotly.express as px

# Gráfico
fig = px.scatter(gols, x='gols_1', y='gols_2',
                 title='Dispersão de Gols Feitos vs Gols Sofridos (Plotly)',
                 labels={'gols_1': 'Gols Marcados (gols_1)', 'gols_2': 'Gols Sofridos (gols_2)'},
                 opacity=0.6)
fig.show()
````

#### Q4.

Apresente um gráfico de barras com o top 10 países que mais participaram de copas do mundo, onde no eixo x devem estar o nome dos países e no eixo y a contagem de participações. Você deve separar a contagem de participações em copas femininas e masculinas, empilhando as barras de cada informação.

No exemplo de barras empilhadas da galeria do matplotlib, imagine que a parte azul são as participações do país em copas masculinas, e em laranja as participações femininas:
* https://matplotlib.org/stable/gallery/lines_bars_and_markers/bar_stacked.html

Você deve implementar essa visualização nas três bibliotecas que vimos:
* Matplotlib
* Seaborn
* Plotly

Garanta que o gráfico tenha pelo menos os atributos:
* título
* rótulos de dimensão.
* legenda

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Dados preparados previamente
participacao = (
    pd.concat([
        wc[['time_1', 'copa', 'ano']],
        wc[['time_2', 'copa', 'ano']].rename(columns={'time_2': 'time_1'})
    ])
    .drop_duplicates()
    .groupby(['time_1', 'copa'])
    .count()
    .reset_index()
    .rename(columns={'ano': 'num_copas', 'time_1': 'país'})
)

# Pegando o top 10 países que mais participaram no total
participacao_total = participacao.groupby('país')['num_copas'].sum().sort_values(ascending=False).head(10)
participacao_top10 = participacao[participacao['país'].isin(participacao_total.index)]

# Separando os dados para o gráfico
pivot = participacao_top10.pivot(index='país', columns='copa', values='num_copas').fillna(0)
````

````python
#### solução com matplotlib
# Cores para cada copa
cores = ['orange', 'blue']

# Plot do gráfico com Matplotlib
fig, ax = plt.subplots(figsize=(10, 6))

# Barras empilhadas
ax.bar(pivot.index, pivot['Masculina'], label='Masculina', color=cores[1])
ax.bar(pivot.index, pivot['Feminina'], bottom=pivot['Masculina'], label='Feminina', color=cores[0])

# Atributos
ax.set_title('Top 10 Países com Mais Participações em Copas do Mundo (Matplotlib)', fontsize=14)
ax.set_xlabel('País')
ax.set_ylabel('Número de Participações')
ax.legend()
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
````

````python
#### solução com seaborn
# Transformar os dados para o formato long
pivot_reset = pivot.reset_index().melt(id_vars='país', value_vars=['Masculina', 'Feminina'], var_name='copa', value_name='num_copas')

# Gráfico do Seaborn
plt.figure(figsize=(10,6))
sns.barplot(data=pivot_reset, x='país', y='num_copas', hue='copa')
plt.title('Top 10 Países com Mais Participações em Copas do Mundo (Seaborn)', fontsize=14)
plt.xlabel('País')
plt.ylabel('Número de Participações')
plt.legend(title='Copa')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
````

````python
#### solução com plotly
# Gráfico do Plotly
fig = px.bar(
    pivot_reset,
    x='país',
    y='num_copas',
    color='copa',
    title='Top 10 Países com Mais Participações em Copas do Mundo (Plotly)',
    labels={'num_copas': 'Número de Participações'},
    text_auto=True
)

# Ajustando o layout
fig.update_layout(barmode='stack', xaxis_title='País', yaxis_title='Número de Participações')
fig.show()
````

#### Q5.

Vamos fazer um compilado com as estatísticas históricas de copas do mundo!

Com a biblioteca de sua preferência você deve criar 4 subplots organizados em um grid de 2 linhas e 2 colunas. Eles devem conter os seguintes gráficos:
* Linha 1, coluna 1: Gráfico de barras com a quantidade de jogos que aconteceram por ano
* Linha 1, coluna 2: Gráfico de área (referências a seguir) com o total de gols por ano, separando as informações de `gols_1` e `gols_2` para distinguir gols em casa e do time visitante.
* Linha 2, coluna 1: Gráfico de área com o total de cartões por ano, separando as informações de cartões amarelos e cartões vermelhos, mas agregando cartões do time 1 ou time 2. Ou seja, uma área com `cartao_amarelo_1 + cartao_amarelo_2` e outra área com `cartao_vermelho_1 + cartao_vermelho_2`.
* Linha 2, coluna 2: Gráfico de barras com o total de gols contra por ano, somando `gols_contra_1` e `gols_contra_2`.

Referências sobre gráfico de área
* Matplotlib: https://matplotlib.org/stable/gallery/lines_bars_and_markers/stackplot_demo.html#sphx-glr-gallery-lines-bars-and-markers-stackplot-demo-py
* Pandas + Matplotlib: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.area.html
* Seaborn: https://seaborn.pydata.org/generated/seaborn.objects.Area.html
* Plotly: https://plotly.com/python/filled-area-plots/

````python
# Imports
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Lendo o CSV
wc = pd.read_csv('wc_formatado.csv')

colunas_numericas = [
    'gols_1', 'gols_2',
    'cartao_amarelo_1', 'cartao_amarelo_2',
    'cartao_vermelho_1', 'cartao_vermelho_2',
    'gols_1_contra', 'gols_2_contra'
]

for coluna in colunas_numericas:
    wc[coluna] = pd.to_numeric(wc[coluna], errors='coerce')

# Configurações do gráfico
plt.style.use('seaborn-v0_8-darkgrid')
fig, axs = plt.subplots(2, 2, figsize=(18, 10))
fig.suptitle('Estatísticas Históricas das Copas do Mundo', fontsize=18)

# Quantidade de jogos por ano
jogos_por_ano = wc.groupby('ano').size()
axs[0,0].bar(jogos_por_ano.index, jogos_por_ano.values, color='royalblue')
axs[0,0].set_title('Quantidade de Jogos por Ano')
axs[0,0].set_xlabel('Ano')
axs[0,0].set_ylabel('Nº de Jogos')

# Total de gols por ano (área)
gols_por_ano = wc.groupby('ano')[['gols_1', 'gols_2']].sum()
axs[0,1].stackplot(
    gols_por_ano.index,
    gols_por_ano['gols_1'],
    gols_por_ano['gols_2'],
    labels=['Gols Time 1', 'Gols Time 2'],
    colors=['lightcoral', 'mediumseagreen']
)
axs[0,1].set_title('Total de Gols por Ano')
axs[0,1].set_xlabel('Ano')
axs[0,1].set_ylabel('Nº de Gols')
axs[0,1].legend()

# Total de cartões por ano (área)
cartoes_por_ano = wc.groupby('ano')[['cartao_amarelo_1', 'cartao_amarelo_2', 'cartao_vermelho_1', 'cartao_vermelho_2']].sum()
cartoes_amarelos = cartoes_por_ano['cartao_amarelo_1'] + cartoes_por_ano['cartao_amarelo_2']
cartoes_vermelhos = cartoes_por_ano['cartao_vermelho_1'] + cartoes_por_ano['cartao_vermelho_2']
axs[1,0].stackplot(
    cartoes_por_ano.index,
    cartoes_amarelos,
    cartoes_vermelhos,
    labels=['Cartões Amarelos', 'Cartões Vermelhos'],
    colors=['gold', 'red']
)
axs[1,0].set_title('Total de Cartões por Ano')
axs[1,0].set_xlabel('Ano')
axs[1,0].set_ylabel('Nº de Cartões')
axs[1,0].legend()

# Total de gols contra por ano (barras)
gols_contra_por_ano = wc.groupby('ano')[['gols_1_contra', 'gols_2_contra']].sum().sum(axis=1)
axs[1,1].bar(gols_contra_por_ano.index, gols_contra_por_ano.values, color='darkorange')
axs[1,1].set_title('Total de Gols Contra por Ano')
axs[1,1].set_xlabel('Ano')
axs[1,1].set_ylabel('Nº de Gols Contra')

# Ajustando layout
plt.tight_layout()
plt.subplots_adjust(top=0.92)  # espaço pro título principal
plt.show()
````
